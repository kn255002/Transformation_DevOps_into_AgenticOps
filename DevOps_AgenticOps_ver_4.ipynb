````python
# Let consider the file is part of module: src/models/llm_factory.py
#
from typing import Literal
from langchain_core.language_models import BaseChatModel
import getpass
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_anthropic import ChatAnthropic
import os
#
#
LLMProvider = Literal["gemini", "anthropic"]
class LLMFactory:
    """
    Factory class to create LLM instances based on choice (Supports Gemini and Anthropic Claude).
    """
    supported_llm = ["gemini", "anthropic"]
    @staticmethod
    def create_llm(
        provider: LLMProvider,
        temperature: float = 0,
        **kwargs
    ) -> BaseChatModel:
        """
        Create and return configured LLM instance.

        Args:
            provider=single LLM (i.e. "gemini" or "anthropic")
            temperature= (0 for deterministic)
            **kwargs: Additional parameters

        Returns:
            Configured LLM instance

        Raises:
            ValueError: If provider is unsupported or API key missing
        """
        if provider not in LLMFactory.supported_llm:
            raise ValueError(
                f"Unsupported provider: {provider}. "
                f"Choose from: {LLMFactory.supported_llm}"
            )

        if provider == "gemini":
          return LLMFactory.create_gemini(temperature, **kwargs)
        elif provider == "anthropic":
          return LLMFactory.create_anthropic(temperature, **kwargs)

    @staticmethod
    def create_gemini(temperature: float, **kwargs) -> ChatGoogleGenerativeAI:
        """Create Google Gemini instance."""
        #api_key = kwargs.get("api_key") or os.getenv("GOOGLE_API_KEY")
        api_key = getpass.getpass("Enter API key for GOOGLE: ")

        if not api_key:
            raise ValueError(
                "Google API key not found. Set GOOGLE_API_KEY environment variable "
                "or pass api_key parameter."
            )

        model_name = kwargs.get("model", "gemini-2.5-flash")
        #model_name = kwargs.get("model", "gemini-3-pro-preview")
        #model_name = kwargs.get("model", "gemini-2.5-flash-preview-09-2025")
        return ChatGoogleGenerativeAI(
            model=model_name,
            temperature=temperature,
            google_api_key=api_key,
            convert_system_message_to_human=True,  # Gemini compatibility
            max_retries=kwargs.get("max_retries", 3),
        )
    @staticmethod
    def create_anthropic(temperature: float, **kwargs) -> ChatAnthropic:
        """Create Anthropic Claude instance."""
        #api_key = kwargs.get("api_key") or os.getenv("ANTHROPIC_API_KEY")
        api_key = getpass.getpass("Enter API key for ANTHROPIC: ")

        if not api_key:
            raise ValueError(
                "Anthropic API key not found. Set ANTHROPIC_API_KEY environment variable "
                "or pass api_key parameter."
            )
        model_name = kwargs.get("model", "claude-sonnet-4-5-20250929")

        return ChatAnthropic(
            model=model_name,
            temperature=temperature,
            anthropic_api_key=api_key,
            max_retries=kwargs.get("max_retries", 3),
            max_tokens=kwargs.get("max_tokens", 4096),
        )

    @staticmethod
    def get_default_provider() -> LLMProvider:
        """Get default provider from environment or fallback to Gemini."""
        provider = os.getenv("LLM_PROVIDER", "gemini").lower()

        if provider not in LLMFactory.supported_llm:
            print(f"вљ пёЏ  Invalid LLM_PROVIDER in .env: {provider}. Using 'gemini'")
            return "gemini"

        return provider  # type: ignore
````

````python
# Let consider the file is part of module: src/models/schemas.py
#

"""
File: src/models/schemas.py
Complete data modeling for the entire application.
"""

from typing import List, Dict, Optional, Literal
from pydantic import BaseModel, Field, field_validator
from datetime import datetime
import re

from pydantic import BaseModel, Field, field_validator
from datetime import datetime

# ============================================================================
# Core Task Models
# ============================================================================

# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # CLASS # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# class: DevOpsTaskSchemas maintains structure of DevOps task:
# Like Taskid/action/resourceType/resourceName/parameters/dependencies/tool and priority.
# Validate action must be in lowercase only.
#
class DevOpsTaskSchemas(BaseModel):
    """ Structured representation of a DevOps task. """

    task_id: Optional[int] = Field(None, description="Auto-assigned task identifier")
    action: str = Field( description="Action to perform: create_user, delete_user, install_package,create_s3_bucket, provision_ec2, etc.")
    resource_type: str = Field(description="Type of resource: user, package, service, file, s3_bucket, ec2_instance")
    resource_name: str = Field(description="Name of the resource (e.g., username, package name)")
    target_hosts: List[str] = Field(default_factory=list, description="List of target machines (for Ansible) ")
    parameters: Dict[str, str] = Field(default_factory=dict, description="Additional parameters (shell, groups, state, etc.)")
    dependencies: List[int] = Field(default_factory=list, description="Task IDs this task depends on")
    tool: Literal["ansible"] = Field(default="ansible", description="Automation tool to use")
    priority: int = Field(default=5, ge=1, le=10, description="Task priority (1=highest)")
    estimated_duration_seconds: int = Field(default=60, description="Expected execution time")

    #Launching decorator 'field_validator' to take a function i.e. 'validate_action' as an arugment.
    #Whenever a new instance of the class created, it automatically calls this function (validate_action) against the value of particular 'action' field to check, if the data for the particular given field is valid before it gets assigned.

    @field_validator("action")
    def validate_action(cls, v):
        """Ensure action follows naming convention."""
        if not re.match(r'^[a-z_]+$', v):
            raise ValueError("Action must be lowercase with underscores only")
        return v

# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # CLASS # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# class: CaseSchema maintains possible structure/elements of a case/incident
# Validate Incident/case must match pattern: CASE-12345, cid_05128408, or PROJECT-123.
#
class CaseSchema(BaseModel):
    """Complete case document structure."""
    case_id: str = Field(description="Unique case identifier (e.g., CASE-12345, Tkt_0512840814)")
    description: str = Field(description="Brief case description")
    tasks: List[DevOpsTaskSchemas] = Field(description="List of tasks to execute")
    priority: Literal["low", "normal", "high", "critical"] = Field(default="normal")
    requested_by: Optional[str] = Field(None, description="Requester email/username")
    created_at: datetime = Field(default_factory=datetime.now)
    target_environment: Literal["on-prem", "aws", "hybrid"] = Field(
        default="on-prem",
        description="Deployment target"
    )
    #Launching decorator 'field_validator' to take a function i.e. 'validate_case_id' as an arugment.
    #Whenever a new instance of the class created, it automatically calls this function (validate_case_id) against the value of particular 'case_id' field to check, if the data for the particular given field is valid before it gets assigned.

    @field_validator("case_id")
    def validate_case_id(cls, v):
        """Ensure case ID follows expected patterns."""
        if not re.match(r'^(CASE-\d+|CID_\d+|TEF_\d+|[A-Z]+-\d+)$', v):
            raise ValueError("Case ID must match pattern: CASE-12345, TEF_0512840814, CID_0512840814, or PROJECT-123")
        return v


# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # CLASS # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# class: GeneratedCode is generating contents for Ansible playbook).
#
class GeneratedCode(BaseModel):
    """Maintains structure for generated (Ansible) code."""
    task_id: int
    tool: Literal["ansible"]
    code_content: str = Field(description="Generated playbook or .tf file content")
    file_name: str = Field(description="Suggested filename")
    variables: Dict[str, str] = Field(default_factory=dict, description="Required variables")
    validation_status: Literal["pending", "valid", "invalid"] = Field(default="pending")
    validation_errors: List[str] = Field(default_factory=list)

# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # CLASS # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# class: AnsiblePlaybook structure only 'Ansible playbook' representation.
#
class AnsiblePlaybook(BaseModel):
    """Structured Ansible playbook representation."""
    playbook_name: str
    target_hosts: List[str]
    become_required: bool = True
    tasks: List[Dict] = Field(description="List of Ansible task dictionaries")
    variables: Dict[str, str] = Field(default_factory=dict)

# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # CLASS # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# class: ExecutionResult maintains Resultset of (Ansible) single task.
#
class ExecutionResult(BaseModel):
    """Result of a single task execution."""

    task_id: int
    case_id: str
    tool: Literal["ansible"]
    status: Literal["success", "failed", "skipped", "timeout"] = Field(default="failed")
    start_time: datetime
    end_time: Optional[datetime] = None
    duration_seconds: Optional[float] = None
    stdout: str = Field(default="")
    stderr: str = Field(default="")
    return_code: int = Field(default=-1)
    error_message: Optional[str] = None

    # Tool-specific metadata
    ansible_stats: Optional[Dict] = Field(None, description="Ansible play recap (ok, changed, unreachable, failed)")

    def calculate_duration(self):
        """Calculate execution duration."""
        if self.end_time:
            self.duration_seconds = (self.end_time - self.start_time).total_seconds()

# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # CLASS # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# class: CaseExecutionReport maintains report for a complete incident/case.
#
class CaseExecutionReport(BaseModel):
    """Comprehensive execution report for entire case."""

    case_id: str
    execution_id: str = Field(description="Unique execution run ID (e.g., 2025-10-24_CASE-12345)" )
    start_time: datetime
    end_time: Optional[datetime] = None
    total_tasks: int
    successful_tasks: int = 0
    failed_tasks: int = 0
    skipped_tasks: int = 0
    overall_status: Literal["success", "partial", "failed"] = Field(default="failed")
    task_results: List[ExecutionResult] = Field(default_factory=list)
    summary: str = Field(default="")
    recommendations: List[str] = Field(default_factory=list)
    artifacts_path: str = Field(description="Directory containing logs and code")

    def calculate_summary(self):
        """Auto-generate summary based on results."""
        if self.failed_tasks == 0:
            self.overall_status = "success"
            self.summary = f"вњ… All {self.total_tasks} tasks completed successfully."
        elif self.successful_tasks > 0:
            self.overall_status = "partial"
            self.summary = (
                f"вљ пёЏ  Partial success: {self.successful_tasks}/{self.total_tasks} tasks completed. "
                f"{self.failed_tasks} failed."
            )
        else:
            self.overall_status = "failed"
            self.summary = f"вќЊ All {self.total_tasks} tasks failed."

# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # CLASS # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# class: SSHCredentials maintains SSH connection credentials.
#
class SSHCredentials(BaseModel):
    """SSH connection credentials."""

    username: str
    password: Optional[str] = None
    private_key_path: Optional[str] = None
    port: int = 22

    def validate_auth_method(self):
        """Ensure either password or key is provided."""
        if self.password is None and self.private_key_path is None:
            raise ValueError("Either password or private_key_path must be provided")

# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # CLASS # # # # # # # # # # # # # # # #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# class: SMTPConfig maintains Email configuration setttings.
#
class SMTPConfig(BaseModel):
    """Email notification configuration."""

    host: str = Field(default="smtp.gmail.com")
    port: int = Field(default=587)
    username: str
    password: str
    from_address: str
    to_addresses: List[str]
    use_tls: bool = True
````

````python
# Let consider the file is part of module: src/agents/pdf_extractor.py
#
import pdfplumber
from typing import Literal, Optional

from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.language_models import BaseChatModel
import structlog
#from ..models.llm_factory import LLMFactory
#from ..models.schemas import CaseSchema, DevOpsTaskSchemas

logger = structlog.get_logger(__name__)
#
class PDFTaskExtractor:
    """
    Extracts DevOps tasks from PDF documents using LLM-powered parsing.
    Supports multiple LLM providers (Gemini, Anthropic) and handles various
    PDF formats including scanned documents.
    """

    def __init__( self, llm_provider: Optional[LLMProvider] = None, llm: Optional[BaseChatModel] = None, max_retries: int = 3 ):
        """
        Initialize PDF extractor with LLM.
        Args:
            llm_provider: LLM provider name ("gemini" or "anthropic")
            llm: Pre-configured LLM instance (overrides llm_provider)
            max_retries: Max retry attempts for LLM calls
        """
        # Import necessary classes from __main__
        from __main__ import CaseSchema, DevOpsTaskSchemas, LLMFactory, LLMProvider, BaseChatModel

        if llm: # If user passed a ready-made LLM instance, then use this one.
            self.llm = llm
        else:   #Otherwise, use llm_provider or ask from LLMFactory to create default one.
            provider = llm_provider or LLMFactory.get_default_provider()
            self.llm = LLMFactory.create_llm(provider, temperature=0)

        self.parser = PydanticOutputParser(pydantic_object=CaseSchema)
        self.max_retries = max_retries  #Max retry attempts for LLM calls
        logger.info(
            "PDF_extractor_initialized",
            llm_provider=type(self.llm).__name__,
            max_retries=max_retries
        )

    # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
    # Method-1: To extract raw text from PDF using pdfplumber.
    #
    def extract_text_from_pdf(self, pdf_path: str) -> str:
        """
        Extract raw text from PDF using pdfplumber.
        ---
        Args: pdf_path: Path to PDF file
        Returns:  Extracted text content
        Raises: FileNotFoundError: If PDF doesn't exist
        Exception: If PDF is corrupted or unreadable
        """
        logger.info("PDF_open_read_attempt", pdf_path=pdf_path)
        try:
            text_content = []
            with pdfplumber.open(pdf_path) as pdf:
                total_pages = len(pdf.pages)
                #logger.debug("pdf_opened", total_pages=total_pages)

                for page_num, page in enumerate(pdf.pages, 1):
                    text = page.extract_text()
                    if text:
                        text_content.append(text)
                        #logger.debug( "Page_extracted", page_num=page_num, char_count=len(text))

            full_text = "\n".join(text_content)

            if not full_text.strip():
                logger.warning("PDF_empty", pdf_path=pdf_path)
                raise ValueError(f"No text content found in {pdf_path}")
            logger.info("PDF_opened_read_successful", total_chars=len(full_text), total_pages=len(text_content))
            return full_text
        except FileNotFoundError:
            logger.error("PDF_not_found", pdf_path=pdf_path)
            raise
        except Exception as e:
            logger.error("PDF_processing_failed", pdf_path=pdf_path, error=str(e))
            raise
    # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
    # Method-2: Develop chain of tasks and executing them
    # i)  Taks-1: Prepare Prompt,
    # ii) Taks-2: Select a suitable LLM,
    # iii). Taks-3: Parse/extract structured details.
    # iv).  Execute/Invoke chain.
    #
    def llmParse_text_caseDetails(self, pdf_text: str) -> CaseSchema:
        """
        Use LLM to parse extracted text into structured CaseSchema.
        --
        Args: pdf_text: Raw text extracted from PDF
        Returns:  Structured CaseSchema with tasks and metadata

        Raises:
        Exception: If LLM parsing fails after retries
        """
        logger.info("RAW_text_parsing_attempt", text_length=len(pdf_text))
        prompt = PromptTemplate(
            template="""You are an expert DevOps task analyzer. Parse the following case document and extract structured information.

Case Document:
{document_text}

**Extraction Instructions:**

1. **Case ID**: Look for patterns like CASE-XXXXX, Tkt_XXXXXX, or ticket numbers

2. **Tasks**: Identify ALL DevOps tasks mentioned. For each task:
   - Determine action type (create_user, delete_user, install_package, start_service, create_s3_bucket, provision_ec2, etc.)
   - Identify resource type (user, package, service, file, s3_bucket, ec2_instance, vpc, etc.)
   - Extract resource name
   - Find target machines/hosts or 'aws' for cloud resources
   - Extract any parameters (groups, shell, state, region, instance_type, etc.)

3. **Tool Classification**:
   - **Ansible** (tool: "ansible"): Linux administration tasks
     * User/Group management
     * Package installation/removal
     * Service management (start/stop/restart)
     * File operations (create/modify/delete)
     * Network configuration
     * System configuration

4. **Dependencies**: Determine which tasks must complete before others
   - Example: "Create VPC" must complete before "Launch EC2 in VPC"
   - Example: "Install package" before "Configure service"

5. **Target Environment**:
   - "on-prem": Tasks on physical/VM machines
   - "aws": Tasks in AWS cloud
   - "hybrid": Mix of both

**Common Task Patterns:**

Ansible:
- create_user, delete_user, modify_user
- install_package, remove_package, update_package
- start_service, stop_service, restart_service, enable_service
- create_file, modify_file, delete_file
- configure_firewall, add_dns_entry

{format_instructions}

**Important**:
- Assign sequential task_id starting from 1
- Extract ALL parameters mentioned in the document
- Be thorough - don't miss any tasks

Provide the structured output:""",
            input_variables=["document_text"],
            partial_variables={
                "format_instructions": self.parser.get_format_instructions()
            }
        )

        chain = prompt | self.llm | self.parser

        for attempt in range(1, self.max_retries + 1):
            try:
                logger.debug("llm_parsing_attempt", attempt=attempt)
                result = chain.invoke({"document_text": pdf_text})

                # Assign task IDs if not present
                for idx, task in enumerate(result.tasks, 1):
                    if task.task_id is None:
                        task.task_id = idx
                logger.info(
                    "RAW_text_parsed_successfully",
                    case_id=result.case_id,
                    task_count=len(result.tasks)
                )
                return result

            except Exception as e:
                logger.warning(
                    "llm_parsing_failed",
                    attempt=attempt,
                    max_retries=self.max_retries,
                    error=str(e)
                )
                if attempt == self.max_retries:
                    logger.error("llm_parsing_exhausted", error=str(e))
                    raise
        raise Exception("LLM parsing failed after all retries")

    # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
    # Method-3: 'validate_case_document' actually, validate extracted case document for completeness and correctness.
    #
    def validate_case_document(self, case: CaseSchema) -> tuple[bool, list[str]]:
        """
        Validate extracted each task completeness and correctness.
        1). Target hosts must be given.
        ---
        Args: case: Extracted CaseSchema
        Returns:  Tuple of (is_valid, list_of_issues)
        """
        issues = []
        # Check for tasks
        if not case.tasks:
            issues.append("No tasks found in case document")

        # Validate each task
        for idx, task in enumerate(case.tasks, 1):
            if not task.target_hosts:
                issues.append(f"Task {idx} ({task.action}) has no target hosts.")

            # Check dependency validity
            for dep_id in task.dependencies:
                if dep_id >= task.task_id:
                    issues.append(

                                  f"Task {idx} has invalid dependency on task {dep_id} "
                        "(dependencies must reference earlier tasks)"
                    )
        is_valid = len(issues) == 0
        if is_valid:
            logger.info("case_validation_passed", case_id=case.case_id)
        else:
            logger.warning(
                "Case_validation_failed",
                case_id=case.case_id,
                issue_count=len(issues)
            )
        return is_valid, issues
````

````python
from __main__ import CaseSchema, DevOpsTaskSchemas, LLMFactory, LLMProvider, BaseChatModel, GeneratedCode

"""
File: src/agents/code_generator.py
Code Generator Agent - Generates Ansible playbooks & configs
"""

from typing import Dict, List, Optional
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.language_models import BaseChatModel
import structlog
import yaml
import re

logger = structlog.get_logger(__name__)

class CodeGeneratorAgent:
    """
    Generates production-ready Ansible playbooks and configurations
    based on extracted DevOps tasks.
    """

    def __init__(self, llm: Optional[BaseChatModel] = None):
        """
        Initialize playbook code generator with LLM.

        Args:
            llm: Pre-configured LLM instance (uses default if None)
        """
        self.llm = llm or LLMFactory.create_llm(
            LLMFactory.get_default_provider(),
            temperature=0
        )
        logger.info("Playbook_generator_initialized", llm=type(self.llm).__name__)

    def generate_ansible_playbook(self, task: DevOpsTaskSchemas) -> GeneratedCode:
        """
        Generate Ansible playbook for a single task.

        Args:
            task: DevOpsTaskSchemas to convert to Ansible

        Returns:
            GeneratedCode with playbook YAML
        """
        logger.info(
            "generating_ansible_playbook",
            task_id=task.task_id,
            action=task.action
        )

        prompt = ChatPromptTemplate.from_template("""
You are an expert Ansible developer or engineer. Before generating an Ansible playbook,
make sure you write clean, strictly valid YAML, production-ready, syntax checked, and following all mentioned requirements, and rules for this task:

**Task Details:**
- Action: {action}
- Resource Type: {resource_type}
- Resource Name: {resource_name}
- Target Hosts: {target_hosts}
- Parameters: {parameters}

**Requirements:**
1. Clean and strictly valid YAML
2. Include error handling (failed_when, ignore_errors where appropriate)
3. Use descriptive variable names
4. Use meaningful task names
5. Add few and only ncecessary tags for selective execution
6. Include pre-flight validation task(s), as system should handle only Ubuntu
7. Use variables for reusability
8. Add handlers if needed (e.g., for service restarts)
9. Follow Ansible best practices
10. If its necessary to explain a complex logic then comments must begin with '#'.

**Example Actions:**
- create_user: Use ansible.builtin.user module with proper parameters
- install_package: Use ansible.builtin.package or ansible.builtin.yum/apt
- start_service: Use ansible.builtin.systemd or ansible.builtin.service
- create_file: Use ansible.builtin.copy or ansible.builtin.template

**Output Requirements:**
- Provide clean, and strictly valid YAML (no markdown, no explanations)
- Start with valid playbook structure (name, hosts, become, tasks)
- Use and prefer YAML list and not a comma-separated string
- Ansible module or plugin or fully qualified collection name should be ending with single ":"
- use Ansible's Jinja2 syntax {{ variable_name }} where appropriate for variables substitution and always wrap the entire string in quotes
- In dictionary-like structures, colons are expected, but the value part should be quoted or formatted accordingly

**Strict rules:**
1. When using Ansible modules (like getent, stat, package or service), Do NOT assume a standard shell return code (rc).
2. Prefer Ansible facts or module behavior over shell commands.
3. For existence checks:
   - Prefer idempotent modules (e.g., user, package, file) instead of manual pre-checks.
   - Check the official Ansible documentation for the specific module's return structure.
   - If getent is used, handle missing entries by setting failed_when: false and checking returned facts safely.

Generate the valid playbook now:""")

        response = self.llm.invoke(
            prompt.format(
                action=task.action,
                resource_type=task.resource_type,
                resource_name=task.resource_name,
                target_hosts=", ".join(task.target_hosts),
                parameters=task.parameters
            )
        )

        code_content = self.extract_yaml_from_response(response.content)

        # Validate YAML syntax
        validation_status, errors = self.validate_ansible_yaml(code_content)

        generated = GeneratedCode(
            task_id=task.task_id,
            tool="ansible",
            code_content=code_content,
            file_name=f"playbook_task_{task.task_id}_{task.action}.yml",
            variables=task.parameters,
            validation_status=validation_status,
            validation_errors=errors
        )

        logger.info(
            "ansible_playbook_generated",
            task_id=task.task_id,
            validation_status=validation_status,
            lines=len(code_content.splitlines())
        )

        return generated

    def generate_code_for_tasks(
        self,
        tasks: List[DevOpsTaskSchemas]
    ) -> Dict[int, GeneratedCode]:
        """
        Generate code for multiple tasks.

        Args:
            tasks: List of DevOpsTaskSchemas objects

        Returns:
            Dictionary mapping task_id to GeneratedCode
        """
        logger.info("generating_code_for_multiple_tasks", task_count=len(tasks))

        generated_codes = {}

        for task in tasks:
            try:
                if task.tool == "ansible":
                    code = self.generate_ansible_playbook(task)
                else:
                    logger.warning(
                        "unsupported_tool",
                        task_id=task.task_id,
                        tool=task.tool
                    )
                    continue

                generated_codes[task.task_id] = code

            except Exception as e:
                logger.error(
                    "code_generation_failed",
                    task_id=task.task_id,
                    error=str(e)
                )

        logger.info(
            "code_generation_complete",
            successful=len(generated_codes),
            total=len(tasks)
        )

        return generated_codes

    # ========================================================================
    # Helper Methods
    # ========================================================================

    def extract_yaml_from_response(self, response: str) -> str:
        """Extract YAML content from LLM response (removes markdown fences)."""
        # Remove markdown code fences
        pattern = r"```(?:yaml|yml)?\n(.*?)```"
        match = re.search(pattern, response, re.DOTALL)

        if match:
            return match.group(1).strip()

        # If no fence, return as-is (assume entire response is YAML)
        return response.strip()

    def validate_ansible_yaml(self, yaml_content: str) -> tuple[str, List[str]]:
        """
        Validate Ansible YAML syntax and attempt to correct using LLM.

        Returns:
            Tuple of (status, list_of_errors)
        """
        errors = []
        corrected_yaml_content = yaml_content # Start with original

        # --- LLM Correction Logic ---
        correction_prompt = ChatPromptTemplate.from_template("""
        You are an expert YAML syntax corrector.
        The following YAML content has syntax errors. Please correct it to be valid YAML,
        while maintaining the original intent and structure as much as possible.
        Only provide the corrected YAML content, without any additional explanations or markdown fences.

        Original YAML with potential errors:
        {original_yaml_content}

        Corrected YAML:""")

        try:
            llm_response = self.llm.invoke(correction_prompt.format(original_yaml_content=yaml_content))
            corrected_yaml_content = self.extract_yaml_from_response(llm_response.content)
            if corrected_yaml_content != yaml_content:
                logger.info("ansible_yaml_corrected_by_llm", original_length=len(yaml_content), corrected_length=len(corrected_yaml_content))
            else:
                logger.debug("ansible_yaml_llm_no_change", original_length=len(yaml_content))
        except Exception as e:
            logger.warning("llm_yaml_correction_failed", error=str(e))
            errors.append(f"LLM YAML correction failed: {str(e)}. Proceeding with original YAML.")
            # If LLM correction fails, we proceed with the original content for validation

        # --- Original Validation Logic (now using corrected_yaml_content) ---
        try:
            data = yaml.safe_load(corrected_yaml_content)

            # Check basic structure
            if not isinstance(data, list):
                errors.append("Playbook must be a list of plays")
            elif len(data) == 0:
                errors.append("Playbook is empty")
            else:
                # Check first play structure
                play = data[0]
                if "name" not in play:
                    errors.append("Play missing 'name' field")
                if "hosts" not in play:
                    errors.append("Play missing 'hosts' field")
                if "tasks" not in play:
                    errors.append("Play missing 'tasks' field")

            status = "valid" if len(errors) == 0 else "invalid"

        except yaml.YAMLError as e:
            errors.append(f"YAML syntax error (even after LLM attempt): {str(e)}")
            status = "invalid"

        return status, errors


````

````python
import os
from pathlib import Path
from datetime import datetime
from typing import Optional, Dict, Any
import structlog

from __main__ import CaseSchema, DevOpsTaskSchemas, ExecutionResult, LLMFactory, LLMProvider, BaseChatModel, GeneratedCode
import ansible_runner
from datetime import datetime
import os

class AnsibleExecutor:
    def __init__(self, case_id, base_dir='./runs'):
      self.case_id = case_id
      self.run_dir = base_dir #Considering default *.ini path.

    def execute_playbook(self, inventory_path: str, playbook_path: str, extra_vars: Optional[Dict[str, Any]] = None):
        if not os.path.isabs(inventory_path):
          inventory_path = os.path.join(self.run_dir, inventory_path)
        if not os.path.isabs(playbook_path):
          playbook_path = os.path.join(self.run_dir, playbook_path)

        run_kwargs = {
            "private_data_dir": self.run_dir,
            "playbook": playbook_path,
            "inventory": inventory_path,
            "extravars": extra_vars or {},
            "quiet": False
        }

        # No need to set ANSIBLE_SSH_PRIVATE_KEY_FILE here, as i tested such parameter if its part of inventory file then chances of error are reduced.
        # And chmod is handled before calling this executor.

        # Execute
        runner = ansible_runner.run(**run_kwargs)

        # Capture results
        return {
            "status": runner.status,  # successful, failed, unreachable
            "rc": runner.rc,
            "stats": runner.stats,
            "events": [event for event in runner.events],
            "stdout": runner.stdout.read()
        }
````
